```python
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
sample_size=[2**i for i in range(5, 14)]
accuracy_function_1 = [0.9855, 0.9840, 0.9858, 0.9912, 0.9922, 0.9935, 0.9945, 0.9959, 0.9989]
accuracy_function_2 = [0.9952, 0.9869, 0.9790, 0.9829, 0.9504, 0.9523, 0.9759, 0.9802, 0.9871]
accuracy_jacobian_1 = [0.0247, 0.0128, 0.1046, 0.2974, 0.3448, 0.5014, 0.5086, 0.7956, 0.8473]
accuracy_jacobian_2 = [0.5866, -0.4354, 0.6390, 0.6848, 0.4234, 0.6393, 0.8151, 0.9323, 0.9871]
```

```python
plt.figure(figsize=(10, 5))
plt.plot(sample_size, accuracy_function_1, label='L2- Norm', marker='o')

plt.plot(sample_size, accuracy_function_2, label='L2- Norm with Jacobian', marker='.')


plt.xlabel('Sample Size')
plt.ylabel('Accuracy')
plt.title('Accuracy vs Sample Size')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(sample_size, accuracy_function_1, label='M2 -Norm', marker='o')
plt.plot(sample_size, accuracy_jacobian_1, label='F2 -Norm', marker='*')
plt.plot(sample_size, accuracy_function_2, label='M2 -Norm with Jacobian', marker='.')
plt.plot(sample_size, accuracy_jacobian_2, label='F2 -Norm with Jacobian', marker='s')

plt.xlabel('Sample Size')
plt.ylabel('Accuracy')
plt.title('Accuracy vs Sample Size')
plt.legend()
plt.grid(True)
plt.show()
```
